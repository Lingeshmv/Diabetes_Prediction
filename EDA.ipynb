## EDA

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the dataset
diabetes_df = pd.read_csv('./Data/diabetes.csv')
```

```python
# Check the shape of the dataset
print(diabetes_df.shape)
```

```python
# Check the first 5 rows of the dataset
print(diabetes_df.head())
```

```python
# Check the last 5 rows of the dataset
print(diabetes_df.tail())
```

```python
# Check the data types of the columns
print(diabetes_df.dtypes)
```

```python
# Check for missing values
print(diabetes_df.isnull().sum())
```

```python
# Check the statistical summary of the dataset
print(diabetes_df.describe())
```

```python
# Visualize the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(diabetes_df.corr(), annot=True, cmap='coolwarm')
plt.show()
```

```python
# Visualize the distribution of the target variable
plt.figure(figsize=(8, 6))
sns.countplot(x='Outcome', data=diabetes_df)
plt.show()
```

```python
# Visualize the distribution of the features
diabetes_df.drop('Outcome', axis=1).hist(bins=50, figsize=(20, 15))
plt.show()
```

```python
# Visualize the pairplot of the dataset
sns.pairplot(diabetes_df, hue='Outcome')
plt.show()
```

```python
# Visualize the boxplot of the dataset
diabetes_df.boxplot(by='Outcome', figsize=(20, 15))
plt.show()
```
